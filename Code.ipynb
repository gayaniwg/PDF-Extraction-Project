```python
from PIL import Image
import pytesseract
import fitz
import re

def extract_text(image):
    custom_config = r'--oem 3 --psm 6 -l eng+sin'
    extracted_text = pytesseract.image_to_string(image, config=custom_config)
    extracted_text = re.sub(r'[^a-zA-Z0-9\s\u0D80-\u0DFF.*!&$?අආඇඈඉ	ඊඋඌඍඎඐඑඒඓඔඕඖකඛගඝඞඟචඡජඣඤඥඦටඨඩඪණඬතථදධනඳපඵබභමඹයලවශෂසහළෆ්ාැෑිීුූෘෙේොෝ	ෞෟ]', '', extracted_text)
    return extracted_text

def preprocess_image(image):
    new_width = 15000
    new_height = int((new_width / image.width) * image.height)
    image = image.resize((new_width, new_height), Image.ANTIALIAS)
    return image

pdf_path = "/content/drive/MyDrive/1/Yat_2014_05.pdf"

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    formatted_text_list = []

    for page_number in range(doc.page_count):
        page = doc[page_number]
        image = page.get_pixmap()


        image = preprocess_image(Image.frombytes("RGB", [image.width, image.height], image.samples))

        extracted_text = extract_text(image)
        formatted_text_list.append(extracted_text)

    return formatted_text_list

formatted_text_list = extract_text_from_pdf(pdf_path)

data = {'Page Number': [], 'Text': []}

for page_number, text in enumerate(formatted_text_list):
    lines = text.splitlines()
    non_blank_lines = [line for line in lines if line.strip() != '']
    data['Page Number'].extend([page_number + 1] * len(non_blank_lines))
    data['Text'].extend(non_blank_lines)

df = pd.DataFrame(data)
print(df)
```

```python
import pdfplumber
import re
import pandas as pd

def extract_specific_lines(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        relevant_lines = []
        for page in pdf.pages:
            page_text = page.extract_text()
            lines = [line.strip() for line in page_text.split('\n') if line.strip()]

            for line in lines:
             if re.search(r'\b\d{9}[A-Z]([A-Za-z .ගවගම]*)?\b|\b\d{12}\b|\*\*\*\*\*\*\*\*\*\*', line):
                    line = re.sub(r'^\[\d+\]', '', line)
                    line = re.sub(r'(\d{9}[A-Z])([A-Za-z .ගවගම]*)', r'\1 \2', line)
                    line = re.sub(r'\*\*\*\*\*\*\*\*\*\*(\S)', r'********** \1', line)

                    relevant_lines.append(line.strip())
    return relevant_lines

def process_lines(lines):
    processed_lines = []
    prev_add_line = None
    for line in lines:
        match = re.match(r'(?P<Add_Line>\d{1,4}(?:\/\d+[A-Za-z]*)?)?(?P<Name>[\S\s]*?)'
                         r'(?P<ID>\d{9}[A-Z]|\*{10})(?P<Gender>[\S\s]*?)'
                         r'(?P<No>\d{1,4})$', line)

        if match:
            add_line = match.group('Add_Line')
            name = match.group('Name').strip()
            english_part = re.match(r'^[A-Za-z0-9/]+', name)
            if english_part:
                add_line = f"{add_line} {english_part.group()}"
                name = name[len(english_part.group()):].strip()

            if add_line is None:
                add_line = prev_add_line
            else:
                prev_add_line = add_line

            gender = match.group('Gender').strip()
            if 'ගව' in gender or 'ගම' in gender:
                gender = 'Female'
            elif 'ප' in gender:
                gender = 'Male'

            processed_lines.append({
                'Add.Line': add_line,
                'Name': name,
                'ID': match.group('ID'),
                'Gender': gender,
                'No': match.group('No')
            })
    return processed_lines


def create_table(lines):
    df = pd.DataFrame(lines)
    return df

def extract_and_create_table(pdf_path):
    extracted_lines = extract_specific_lines(pdf_path)
    processed_lines = process_lines(extracted_lines)
    table_df = create_table(processed_lines)
    return table_df

table_df = extract_and_create_table(pdf_path)
print(table_df)
```

```python

import pandas as pd

def validate(id_str: str) -> bool:
    id_length = len(id_str)
    if (id_length == 12 or (id_length == 10 and id_str[:-1].isdigit() and (id_str[-1] == 'v' or id_str[-1] == 'V'))):
        return True
    else:
        return False

def change_id_type(id_str: str) -> str:
    if len(id_str) == 10:
        return '19' + id_str[:-1] + '0' + id_str[-1]
    else:
        return id_str

def get_born_year(id_str: str) -> int:
    return int(id_str[:4])

def get_gender(id_str: str) -> str:
    if int(id_str[4:7]) < 500:
        return "Male"
    else:
        return "Female"

def get_born_date(id_str: str) -> datetime.date:
    born_year = get_born_year(id_str)
    born_day = int(id_str[4:7])
    if born_day > 500:
        born_day -= 500
    date_calculator = datetime.date(2000, 1, 1) + datetime.timedelta(days=born_day - 1)
    born_date = date_calculator.replace(year=born_year)
    return born_date

filtered_df['ID'] = filtered_df['ID'].astype(str)

filtered_df['Valid_ID'] = filtered_df['ID'].apply(validate)
df_valid = filtered_df.copy()

df_valid.loc[df_valid['Valid_ID'] == False, ['ID', 'Gender', 'Birthday', 'Valid_ID']] = [None, None, None, None]

df_valid['Duplicate_ID'] = df_valid['ID']

df_valid.loc[df_valid['Valid_ID'] == True, 'Duplicate_ID'] = df_valid.loc[df_valid['Valid_ID'] == True, 'Duplicate_ID'].apply(change_id_type)

df_valid.loc[df_valid['Valid_ID'] == True, 'Gender'] = df_valid.loc[df_valid['Valid_ID'] == True, 'Duplicate_ID'].apply(get_gender)
df_valid.loc[df_valid['Valid_ID'] == True, 'Birthday'] = df_valid.loc[df_valid['Valid_ID'] == True, 'Duplicate_ID'].apply(get_born_date)

df_valid['Birth_Year'] = df_valid['Birthday'].apply(lambda x: x.year if not pd.isnull(x) else np.nan)
df_valid['Birth_Month'] = df_valid['Birthday'].apply(lambda x: x.month if not pd.isnull(x) else np.nan)
df_valid['Birth_Day'] = df_valid['Birthday'].apply(lambda x: x.day if not pd.isnull(x) else np.nan)

print(df_valid[['ID', 'Duplicate_ID', 'Gender', 'Birthday', 'Birth_Year', 'Birth_Month', 'Birth_Day']])
```
